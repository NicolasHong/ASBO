```python
import pandas as pd
from pathlib import Path
BASE_DIR = Path().absolute()
DIR = BASE_DIR/'Results'

data = pd.read_excel(DIR/'slsqp_Results_Case3_100.xls', usecols='C:E', nrows=100,names=['Objective','Evaluation number', 'Feasible'])
data
```

```python
part1 = data[data['Feasible'] == True]
part2 = data[data['Feasible'] == False]
part2
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Set the font to Times New Roman for all text elements
plt.rcParams['font.family'] = 'Times New Roman'

y = data.loc[:,'Objective']
x = data.loc[:,'Evaluation number']
y1 = part1.loc[:,'Objective']
x1 = part1.loc[:,'Evaluation number']
y2 = part2.loc[:,'Objective']
x2 = part2.loc[:,'Evaluation number']

plt.figure(figsize=(10, 5))
sns.kdeplot(x, y, cmap="Blues", shade=True, cbar=True)

plt.scatter(x2, y2, color='red', label='Infeasible')

plt.scatter(x1, y1, color='blue', label='Feasible')


plt.xscale('log')

plt.xlabel('Objective',fontsize=15)
plt.ylabel('Evaluation number',fontsize=15)

baseline = min(y) 
plt.axhline(baseline, color='green', linestyle='--', label='Minimum')
plt.legend()
plt.savefig('Case 3 Figure 4.png',dpi=500, bbox_inches='tight')

plt.show()
```
