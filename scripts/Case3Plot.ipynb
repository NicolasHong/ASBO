```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

plt.rcParams['font.family'] = 'Times New Roman'

from pathlib import Path
BASE_DIR = Path().absolute().parent
DIR = BASE_DIR/'scripts_res'
DIR
```

```python
file_path1 = DIR/'ASBO_Results_Case1_5.csv' 
file_path2 = DIR/'TuRBO_Results_Case1_5.csv' 
file_path3 = DIR/'CBO_Results_Case1_5.csv' 
file_path4 = DIR/'GA_Results_Case1_5.csv' 
file_path5 = DIR/'PSO_Results_Case1_5.csv' 
file_path6 = DIR/'slsqp_Results_Case1_5.csv'
file_path7 = DIR/'trust-constr_Results_Case1_5.csv' 


# file_path1 = DIR/'ASBO_Results_Case3_100.xls' 
# file_path2 = DIR/'botorch_Results_Case3_100.xls' 
# file_path3 = DIR/'CBO_Results_Case3_100.xls' 
# file_path4 = DIR/'GA_Results_Case3_100.xls' 
# file_path5 = DIR/'PSO_Results_Case3_100.xls' 
# file_path6 = DIR/'slsqp_Results_Case3_100.xls'
# file_path7 = DIR/'trust-constr_Results_Case3_100.xls' 

df = pd.read_csv(file_path1)
data_list1 = [-num for num in df['obj']]
df = pd.read_csv(file_path2)
data_list2 = [-num for num in df['obj']]
df = pd.read_csv(file_path3)
data_list3 = [-num for num in df['obj']]
df = pd.read_csv(file_path4)
data_list4 = df['obj'].tolist() 
df = pd.read_csv(file_path5)
data_list5 = df['obj'].tolist() 
df = pd.read_csv(file_path6)
data_list6 = df['obj'].tolist() 
df = pd.read_csv(file_path7)
data_list7 = df['obj'].tolist() 
```

```python
optimal_value1 = min(data_list1)
regret1 = [result - optimal_value1 for result in data_list1]
optimal_value2 = min(data_list2)
regret2 = [result - optimal_value2 for result in data_list2]
optimal_value3 = min(data_list3)
regret3 = [result - optimal_value3 for result in data_list3]
optimal_value4 = min(data_list4)
regret4 = [result - optimal_value4 for result in data_list4]
optimal_value5 = min(data_list5)
regret5 = [result - optimal_value5 for result in data_list5]
optimal_value6 = min(data_list6)
regret6 = [result - optimal_value6 for result in data_list6]
optimal_value7 = min(data_list7)
regret7 = [result - optimal_value7 for result in data_list7]

#  regret plot
plt.figure(figsize=(10, 6))
plt.plot(regret1)#, marker='o', linestyle='-', color='b'
plt.plot(regret2)
plt.plot(regret3)
plt.plot(regret4)
plt.plot(regret5)
plt.plot(regret6)
plt.plot(regret7)
plt.title('Regret Plot')
plt.xlabel('Iterations')
plt.ylabel('Regret')
plt.grid()
plt.axhline(0, color='red', linestyle='--')
plt.show()
```
