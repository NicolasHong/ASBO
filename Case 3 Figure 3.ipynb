```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

plt.rcParams['font.family'] = 'Times New Roman'

from pathlib import Path
BASE_DIR = Path().absolute()
DIR = BASE_DIR/'Results'

file_path = DIR/'Case 3 Figure 3.xls' 
```

```python
df = pd.read_excel(file_path, sheet_name='a')
iterations = df['Iteration']
objective_values = df['ObjectiveValue']
probabilities = df['Probability']
```

```python
plt.figure(figsize=(12, 4), dpi= 100)
norm = mcolors.Normalize(vmin=0, vmax=1)
sc = plt.scatter(iterations, objective_values, c=probabilities, cmap='viridis', s=50, norm=norm)
cbar = plt.colorbar(sc)
cbar.set_label('Probability of satisfying constraints', fontsize=14, fontweight='bold', color='black', family='Times New Roman', labelpad=10)
plt.xlabel('Iteration', fontsize=12, fontweight='bold', color='black')
plt.ylabel('Objective values', fontsize=12, fontweight='bold', color='black')

plt.gca().xaxis.set_tick_params(labelsize=10, labelcolor='black', width=1)
plt.gca().yaxis.set_tick_params(labelsize=10, labelcolor='black', width=1)

plt.tick_params(direction='in')

plt.ylim((-3, 10))
baseline = 1.06956  # Baseline
plt.axhline(baseline, color='green', linestyle='--', label='Minimum')
plt.legend()
plt.savefig('Case 3 Figure 3a.png',dpi=500, bbox_inches='tight')
plt.show()
```
